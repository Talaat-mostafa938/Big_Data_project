```python
import os
os.environ["HADOOP_USER_NAME"] = "root"
```

```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession\
        .builder\
        .appName('ReadFromHDFS')\
        .master('local[*]')\
        .getOrCreate()
```

```python
file_name = ['customers' , 'orders' , 'order_items' , 'products' ]

base_path = "hdfs://hadoop-namenode:9000/user/hadoop/data/Big_Data/raw_data/"
output_path = "hdfs://hadoop-namenode:9000/user/hadoop/data/Big_Data/parquet/"

for name in file_name:
    path = f'{base_path}{name}.csv'

    df = spark.read.csv(path , header=True , inferSchema=True)
    count = df.count()
    
    if df.count() > 0:
        df.write.mode('overwrite').parquet(f'{output_path}{name}')
    
        print(f"✅ {name}: {count} save in Parquet")

spark.stop()
print("🎉 All files have been successfully converted to Parquet!")
```
